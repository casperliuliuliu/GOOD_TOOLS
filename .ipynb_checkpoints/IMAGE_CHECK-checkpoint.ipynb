```python
from PIL import Image
import os

def get_image_resolution(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
        return width, height

def process_images_in_folder(folder_path, nouse=None ,delete_option = 0):
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
            image_path = os.path.join(folder_path, filename)
            width, height = get_image_resolution(image_path)
            if width <= 300 and height <= 300:
                if delete_option == 1:
                    os.remove(image_path)
                    print(f"Deleted: {filename}")
                else:
                    print(f"Image: {filename}, Resolution: {width} x {height}")


# if __name__ == "__main__":
# for ii in 
folder_path = "E:\\Data\\中國醫腎絲球分割output\\MGN_forCNN"
for ii in os.listdir(folder_path):
    # print(type(ii))
    sub_path = f"{folder_path}\{ii}"
    print(sub_path)
    process_images_in_folder(sub_path, delete_option=1)
# os.listdir(folder_path)
```
