```python
# Importing the library
import psutil
import GPUtil
print("[GPU]")
GPUtil.showUtilization()
# # Getting % usage of virtual_memory ( 3rd field)
print("\n[RAM]")
mem_usage = psutil.virtual_memory()
print(f"Free RAM: {mem_usage.percent}%")
print(f"Total RAM: {mem_usage.total/(1024**3):.2f}G")
print(f"Used RAM: {mem_usage.used/(1024**3):.2f}G")

print(f"\n[CPU]")
per_cpu = psutil.cpu_percent(percpu=True,interval=1)
# For individual core usage with blocking, psutil.cpu_percent(interval=1, percpu=True)
avg = 0
for idx, usage in enumerate(per_cpu):
    avg += usage
    
# calculate average CPU usage
print(f"CORE_avg: {avg/(idx+1)}%\n")

for idx, usage in enumerate(per_cpu):
    print(f"CORE_{idx}: {usage}%")
```

```python
# Imports
from os import get_terminal_size

for process in [psutil.Process(pid) for pid in psutil.pids()]:
    try:
        name = process.name()
        mem = process.memory_percent()
        cpu = process.cpu_percent(interval=0.5)
    except psutil.NoSuchProcess as e:
        print(e.pid, "killed before analysis")
    else:
        print("Name:", name)
        print("CPU%:", cpu)
        print("MEM%:", mem)
    print("-" * get_terminal_size().lines)
```
